# Distribución de probabilidad de una variable aleatoria discreta


Consideremos una variable aleatoria $X$ con rango  $R_X=\{x_1,x_2,x_3,...\}$. 

La funcion función de probabilidad (o función de masa de probabilidad) (PMF) determina la probabilidad que una realización de la misma corresponda a un determinado elemento de su rango. Tenemos entonces que:
$$P(X=x)=P_X(x)$$

$$\sum_{x \in R_X} P_X(x)=1$$

Aqui aprenderemos un método no paramétrico para estimar la probabilidad de un intervalo de una variable aleatoria continua.

```python
%matplotlib inline
from scipy.stats.kde import gaussian_kde
from scipy.stats import norm
from numpy import linspace,hstack
from pylab import plot,show,hist,bar

# creating data with two peaks
sampD1 = norm.rvs(loc=-1.0,scale=1,size=300)
sampD2 = norm.rvs(loc=2.0,scale=0.5,size=300)
samp = hstack([sampD1,sampD2])

# obtaining the pdf (my_pdf is a function!)
my_pdf = gaussian_kde(samp)

# plotting the result
x = linspace(-5,5,20)

bar(x,my_pdf(x),color='red',alpha=.3) # distribution function
plot(x,my_pdf(x),color='k') # distribution function
hist(samp,normed=1,alpha=.3) # histogram
show()
```
