# KDE Kernel Density Estimation

Aqui aprenderemos un método no paramétrico para estimar la probabilidad de un intervalo de una variable aleatoria continua.


Consideremos una variable aleatoria $X$ con rango  $(X)=RX(X)=RX$. La  PMF determina la probabilidad que una realizacion de la variable $X$ corresponda a un elemento de su rango. Tenemos entonces que :

$\sum_{x \in R_X} P_X(x)=1$


```python
%matplotlib inline
from scipy.stats.kde import gaussian_kde
from scipy.stats import norm
from numpy import linspace,hstack
from pylab import plot,show,hist

# creating data with two peaks
sampD1 = norm.rvs(loc=-1.0,scale=1,size=300)
sampD2 = norm.rvs(loc=2.0,scale=0.5,size=300)
samp = hstack([sampD1,sampD2])

# obtaining the pdf (my_pdf is a function!)
my_pdf = gaussian_kde(samp)

# plotting the result
x = linspace(-5,5,100)
plot(x,my_pdf(x),'r') # distribution function
hist(samp,normed=1,alpha=.3) # histogram
show()
```
